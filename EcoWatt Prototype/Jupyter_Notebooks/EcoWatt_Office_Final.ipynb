Importing the Necessary Libraries :-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline , make_pipeline
```

Loading The Data :-

```python
df = pd.read_csv("/content/EcoWatt_Office_Realistic_Updated.csv")
df.head()
```

Replacing The NAn Values with None :-

```python
df.fillna("None", inplace=True)
df.loc[df['AC_Type'] == 'None', 'Monthly_AC_Usage_Hours'] = 0
df.loc[df['Fan_Type'] == 'None', 'Monthly_Fan_Usage_Hours'] = 0
df.loc[df['Refrigerator_Type'] == 'None', 'Refrigerator_Usage_Hrs_Monthly'] = 0
df.loc[df["Lights_Type"] == 'None', 'Monthly_Lights_Usage_Hours'] = 0
df.loc[df["Printer_Type"] == 'None','Monthly_Printer_Usage_Minutes' ] = 0
df.loc[df['Projector_Type'] == 'None','Monthly_Projector_Usage_Hours' ] = 0
```

Dividing Data Into Independent & Dependent Columns :-

```python
X = df[[ 'State', 'City', 'Weather/Season', 'Office_Type',
       'Office_Scale', 'Electricity_Tariff_Type', 'Avg_Working_Hours_Monthly',
       'No_Of_ACs', 'AC_Type', 'Monthly_AC_Usage_Hours', 'No_Of_Fans',
       'Fan_Type', 'Monthly_Fan_Usage_Hours', 'No_Of_Lights', 'Lights_Type',
       'Monthly_Lights_Usage_Hours', 'No_Of_PCs', 'PC_Type',
       'Monthly_PC_Usage_Hours', 'No_Of_Refrigerators', 'Refrigerator_Type',
       'Monthly_Refrigerator_Usage_Hours', 'No_Of_Printers', 'Printer_Type',
       'Monthly_Printer_Usage_Minutes', 'No_Of_Projectors', 'Projector_Type',
       'Monthly_Projector_Usage_Hours']]
y = df["Monthly_Total_Kwh_Consumption"]
```

Data Transformation Task :-

```python
Nominal_cols = ['State', 'City', 'Weather/Season']
Ordinal1_cols = ['Office_Type','Office_Scale','Electricity_Tariff_Type']
Ordinal2_cols = ['AC_Type','Fan_Type','Lights_Type','PC_Type','Refrigerator_Type','Printer_Type','Projector_Type']
Numerical1_cols = ['Avg_Working_Hours_Monthly','Monthly_AC_Usage_Hours','Monthly_Fan_Usage_Hours','Monthly_Lights_Usage_Hours','Monthly_PC_Usage_Hours','Monthly_Refrigerator_Usage_Hours','Monthly_Printer_Usage_Minutes','Monthly_Projector_Usage_Hours']
Numerical2_cols = ['No_Of_ACs','No_Of_Fans','No_Of_Lights','No_Of_PCs','No_Of_Refrigerators','No_Of_Printers','No_Of_Projectors']

Ordinal1_cat = [['IT/Corporate', 'Startup', 'Government_Office'],['Medium', 'Large', 'Small'],['LT-2 (Commercial)']]
Ordinal2_cat  = [['Split AC', 'Inverter AC', 'Window AC', 'None'],['Bajaj', 'Crompton', 'Havells', 'None', 'Polycab'],['LED', 'CFL', 'Tube Light'],['Laptop', 'Desktop'],['None', 'Side-by-Side 4★', 'Double Door 4★', 'Single Door 3★','Double Door 3★', 'Side-by-Side 5★', 'Single Door 5★','Double Door 5★', 'Single Door 4★'],['HP', 'Canon', 'Sony'],['Epson', 'Sony']]


transformer = ColumnTransformer(
    transformers=[
        ("tnf1", OneHotEncoder(drop="first"), Nominal_cols),
        ("tnf2", OrdinalEncoder(categories= Ordinal1_cat), Ordinal1_cols),
        ("tnf3", OrdinalEncoder(categories= Ordinal2_cat), Ordinal2_cols),
        ("tnf4", MinMaxScaler(), Numerical1_cols),
        ("tnf5", MinMaxScaler(), Numerical2_cols)

    ],
    remainder="passthrough"
)
```

```python
from sklearn.pipeline import Pipeline
Pipe = Pipeline([
    ("tnf1",transformer)
])
```

Dividing The Data into Train-Test Splits :-

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8 ,random_state = 42)

X_train_trans = Pipe.fit_transform(X_train)
X_test_trans = Pipe.transform(X_test)
```

Model Buidling And Evaluation :-

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestRegressor(n_estimators=10 , random_state = 42 , max_samples = 0.8)
```

```python
rf.fit(X_train_trans ,y_train)
```

```python
y_pred = rf.predict(X_test_trans)
print(r2_score(y_test,y_pred))
```

Performing Hyper-Paramter Tunning :-

```python
n_estimators = [10,15,20,25]
max_features = [0.75,0.8,0.9,0.95]
max_depth = [4,5,6]
max_samples = [0.7,0.8,0.9]
```

```python
param_grid = {
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth': max_depth ,
    'max_samples':max_samples
}
print(param_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_grid = GridSearchCV(
    estimator = rf ,
    param_grid =param_grid ,
    cv = 10 ,
    verbose = 2,
    n_jobs = -1
)
```

```python
rf_grid.fit(X_train_trans , y_train)
```

```python
rf_grid.best_params_
```

```python
print(rf_grid.best_score_)
```

Building the Best Model :-

```python
RF= RandomForestRegressor( max_depth = 4 , max_features = 0.75 , max_samples = 0.7, n_estimators=10)
```

```python
RF.fit(X_train_trans , y_train)
y_pred = RF.predict(X_test_trans)
print(r2_score(y_test , y_pred))
```

Dumping the Preprocessing Pipe and the Best Model :-

```python
import joblib
# Saving the preprocessing Pipeline
joblib.dump(Pipe ,"Preprocessing_EcoWatt_Office.pkl")

# Saving the Best Model
joblib.dump(RF , "Best_Model_EcoWatt_Office.pkl")
print("Model And Preprocessing Object Saved SuccessFully .")
```
